## Item-based explanation----

Yes, we can explain the recommendations using item-based explanations. Just like Netflix's section of movie you have liked and facebook's Ads you may like, similarly we can give item based explanations that since user have listened to these artists, That's why these new artists are recommended.

Suppose if an artist is recommended to a perticular user, then if we map items around that certain user with which he have already interacted with and similarly if we map neighbour items around the item that was recommended, then both of these sets will comparatively overlap a lot.

## Computing tags_preference

```python
selected_tags = user_feature_mat.columns.values
selected_tags
```

```python
TAG_PREFERENCE = []

for user in selected_users:
    ui_rating_vec = user_item_rating_mat.loc[user][list(item_feature_mat.index.values)]
    user_rating_mean = smaller_ratings_table[smaller_ratings_table['userID'] == user]['ratings'].mean()
    K = 0.05**7
    tag_pref = []
    
    for tag in selected_tags:
        tag_share_vec = item_feature_mat[tag]/sum(item_feature_mat[tag])
        ut_preference = (np.dot(ui_rating_vec, tag_share_vec) + (user_rating_mean * K)) / (np.sum(tag_share_vec) + K)
        tag_pref.append(ut_preference)
        
    TAG_PREFERENCE.append(tag_pref)
```

## Formulation for ranking tags

1. To calculate "TAG_PREFERENCE(u,t)" for a certain userID 'u' and tagID 't', we can first find out the artistIDs that user 'u' have interacted with.
2. Suppose complete set of item is I and interaction set with userID 'u' is I' which is of length 'k'.
3. Based on this I' set, we can get avg. ratings(average of all the ratings given to that perticular artist) for these interacted artists, which will also be a k-length vector.
4. Now, using the item-tag-matrix, we already have the count for tagID 't' over all the artists.

so, Formula for TAG_PREFERENCE(u,t) = { Σ{count(i,t) * Average_Rating(i)} } / Σ{count(i,t) ; over all items i ∈ I'.

### Writing Function for iterative approach

```python
def hit_rate_hist(USERS):
    hit_rate = [0]*6
    for test_user in USERS:
        items_for_user2 = smaller_ratings_table[smaller_ratings_table['userID'] == test_user]['artistID'].unique()
        
        TARGET_ITEM = smaller_ratings_table[smaller_ratings_table['userID'] == test_user]['artistID'].unique()[0]
        TOP5_neighbours = ii_sim_mat[TARGET_ITEM].sort_values(ascending = False)[1:6].index.values
        SEED = TOP5_neighbours[4]
        
        TOP3_recommendations = []
        FLAG = TARGET_ITEM in TOP3_recommendations
        iterations = 0
        while(iterations < 5):
            SEED_neighbours = ii_sim_mat[SEED].sort_values(ascending = False)[1:11].index.values
            
            SEED_neighbours_similarity = []
            for item in SEED_neighbours:
                a = ii_sim_mat[item].sort_values(ascending=False).index[1:21].values
                inters = np.array(list(set(a).intersection(set(items_for_user2))))
                
                temp_sim_score = np.sum(ii_sim_mat[item][inters])/len(inters)
                SEED_neighbours_similarity.append(temp_sim_score)
            
            SEED_neighbours_similarity = np.array(SEED_neighbours_similarity)
            SEED_neighbours_similarity = pd.DataFrame(SEED_neighbours_similarity, index = SEED_neighbours, columns = ['similarity'])
            TOP3_recommendations = SEED_neighbours_similarity.sort_values(by = 'similarity', ascending = False)[0:3].index.values
            
            FLAG = TARGET_ITEM in TOP3_recommendations
            
            if FLAG:
                hit_rate[iterations+1] += 1
                break
                
            first_neighbour = TOP5_neighbours[0]
            SEED = ii_sim_mat[first_neighbour][TOP3_recommendations].sort_values(ascending = False).index.values[0]
            
            iterations = iterations+1
        
    return hit_rate
```

```python
USERS = user_feature_mat.index.values[0:100]
hit_rate = hit_rate_hist(USERS)
hit_rate[0] = len(USERS) - sum(hit_rate)
hit_rate
```
