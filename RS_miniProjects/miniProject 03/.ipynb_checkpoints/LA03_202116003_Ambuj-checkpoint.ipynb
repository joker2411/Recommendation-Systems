### Calculating RMSE

```python
def calculate_RMSE(Y, Y_real):
  error = np.subtract(Y, Y_real)
  squared_error = np.square(error)
  total_squared_error = squared_error.sum()     #### to sum all the values in each row
  total_mean_squared_error = total_squared_error/len(Y)
  total_root_mean_squared_error = np.sqrt(total_mean_squared_error)
  return total_mean_squared_error
```

```python
calculate_RMSE(estimated_ratings, actual_ratings)
```

### Calculating MAE

```python
def calculate_MAE(Y, Y_real):
  error = np.subtract(Y, Y_real)
  absolute_error = np.abs(error)
  total_absolute_error = absolute_error.sum()     #### to sum all the values in each row
  mean_absolute_error = total_absolute_error/len(Y)
  return mean_absolute_error
```

```python
calculate_MAE(estimated_ratings, actual_ratings)
```

## Recommending predictions for users

### A.) Creating the recommended recipe matrix

```python
K = 50

rec_movie_mat = []
rec_movie_sim_score_mat = []
for i in range(len(distinct_users[:4])):
    sim = pd.DataFrame(similarity_matrix[i], columns = ['Similarity'])
    sim.dropna(inplace=True)
    temp_rec_df = sim[sim['Similarity'] > 0.50].sort_values(by = 'Similarity',ascending=False).head(K)
    
    rec_movie_id = list(temp_rec_df.index)
    rec_movie_sim_score = list(temp_rec_df.Similarity)
    
#     print(rec_movie_id)

    rec_movie_mat.append(rec_movie_id)
    rec_movie_sim_score_mat.append(rec_movie_sim_score)
```

```python
len(rec_movie_mat[0])
```

```python
len(rec_movie_sim_score_mat[0])
```

### B.) Calculating evaluation scores

### Calculating precision, recall, f1score

```python
distinct_users[0:4]
```

```python
PRECISION = []
RECALL = []
F1SCORE = []

for i in range(len(distinct_users[0:4])):
    user = distinct_users[i]
    
    rec_set = set(rec_movie_mat[i])
    temp_df = ratings_csv[ratings_csv['userId'] == user]
    actual_liked_set = set(temp_df[temp_df['rating'] >= 2.5]['movieId'].unique())
    
    ret_rel_set = actual_liked_set.intersection(rec_set)
    
    pre = len(ret_rel_set)/len(rec_set)
    reca = len(ret_rel_set)/len(actual_liked_set)
    if(pre+reca != 0):
        f1sc = (2*pre*reca)/(pre+reca)
    else:
        f1sc = 0
    PRECISION.append(pre)
    RECALL.append(reca)
    F1SCORE.append(f1sc)
```

### printing evalution values 

```python
print('Evaluation measures------->')
print()
for i in range(len(distinct_users[0:4])):
    print('Precision value for user with user id {} is : {}'.format(distinct_users[i], PRECISION[i]))
    print('Recall value for user with user id {} is : {}'.format(distinct_users[i], RECALL[i]))
    print('F1score value for user with user id {} is : {}'.format(distinct_users[i], F1SCORE[i]))
    print('============================================================================')
```

### Calculating DCG, IDCG and RR score for each user

```python
DCG_list = list()
IDCG_list = list()
RR_list = list()

for i in range(len(distinct_users[0:4])):
    user = distinct_users[i]
    rec_movie_ids = rec_movie_mat[i]
    x1 = ratings_csv[ratings_csv['userId'] == user]
    x = x1['movieId'].unique()
    intersection_points = set(rec_movie_ids).intersection(set(x))
    y = x1.loc[x1['movieId'].isin(list(intersection_points))]
    
    
    sorted_y = y.sort_values(by = 'rating', ascending = False)
    intersection_movie_ratings = list(sorted_y['rating'].values)

    rel_sc = len(intersection_movie_ratings)-1
    rel_sc_vec = list()

    if len(intersection_movie_ratings) < 1:
#         nDCG_list.append(0)
        continue
        
    last_value = intersection_movie_ratings[0]

    for val in intersection_movie_ratings:
        if val == last_value:
            rel_sc_vec.append(rel_sc)
        else:
            rel_sc = rel_sc-1
            rel_sc_vec.append(rel_sc)
        last_value = val

    sorted_y['rel_score'] = rel_sc_vec
    
    movieId_rel_dict = dict(zip(list(sorted_y['movieId']), list(sorted_y['rel_score'])))
    ordered_movieId_intersection = [a for a in rec_movie_ids if a in intersection_points]
    
    ordered_rel = list()
    for val in ordered_movieId_intersection:
        ordered_rel.append(movieId_rel_dict[val])
        
    ideal_rel = sorted(ordered_rel, reverse = True)
    
    DCG = 0
    IDCG = 0
    RR = 0
    for i in range(1,len(ordered_rel)+1):
        DCG = DCG+ ((2**ordered_rel[i-1] - 1) / math.log2(i+1))
        IDCG = IDCG+ ((2**ideal_rel[i-1] - 1) / math.log2(i+1))
        
    RANK = ordered_rel.index(max(ordered_rel))+1
    RR = 1/RANK
    
    DCG_list.append(DCG)
    IDCG_list.append(IDCG)
    RR_list.append(RR)
```

### Calculating nDCG score over all the users

```python
DCG_final = sum(DCG_list)/len(DCG_list)
IDCG_final = sum(IDCG_list)/len(IDCG_list)
nDCG = DCG_final/IDCG_final

print('DCG Score over all the users : {}'.format(DCG_final))
print('IDCG Score over all the users : {}'.format(IDCG_final))
print('nDCG Score over all the users : {}'.format(nDCG))
```

```python
MRR = sum(RR_list)/len(RR_list)

print('MRR Score over all the users : {}'.format(MRR))
```

```python
TRAIN_df = pd.DataFrame()
TEST_df = pd.DataFrame()

for user in users_involved_in_rating:
    temp_df = useful_dataframe[useful_dataframe['userId'] == user]
    
    testing_data_num = int(temp_df.shape[0] * 0.3)
    TEST_df = pd.concat([TEST_df, temp_df[0:testing_data_num]], axis=0,ignore_index=True)
    TRAIN_df = pd.concat([TRAIN_df, temp_df[testing_data_num:]], axis=0,ignore_index=True)
    
    
###### Making the dataset compatible for training and testing
TRAIN_df.drop(['timestamp'], axis=1, inplace = True)
TRAIN_df.columns = ['user_id', 'item_id','rating']

TEST_df.drop(['timestamp'], axis=1, inplace = True)
TEST_df.columns = ['user_id', 'item_id','rating']

TRAIN_df_X = TRAIN_df[['user_id', 'item_id']]
TRAIN_df_Y = TRAIN_df['rating']

TEST_df_X = TEST_df[['user_id', 'item_id']]
TEST_df_Y = TEST_df['rating']
```
